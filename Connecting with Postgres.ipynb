```python
import  sqlalchemy as sql_al
import pandas as pd
from sqlalchemy import create_engine
from time import time
```

```python

#Read the data file that we want to add it to the data base
df=pd.read_csv("C:\\Users\\user\\Desktop\\Dataengineering\\green_tripdata_2019-09.csv")
```

```python
#Change the type of drop of time and pickup time

df["lpep_pickup_datetime"]= pd.to_datetime(df["lpep_pickup_datetime"])
df["lpep_dropoff_datetime"]= pd.to_datetime(df["lpep_dropoff_datetime"])
```

```python
#Get the Credentials for the database
engine = create_engine(f'postgresql://root:root@localhost:5432/green_taxi')
```

```python
#Connect with the database
engine.connect()
```

```python
#Dividing the dataset into chunks
df_iter = pd.read_csv('data_to_ingest.csv', iterator=True, chunksize=10000)
```

```python
while True: 
    t_start = time()

    df = next(df_iter)

    df.to_sql(name='green_taxi_data', con=engine, if_exists='append')

    t_end = time()

    print('inserted another chunk, took %.3f second' % (t_end - t_start))
```
